```python
from sklearn.model_selection import train_test_split
```

```python
# split into X,y
X = df_new.drop('fuel',axis=1)
y = df_new['fuel']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

#using Sklearn


```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

## 1. Applying Label Encoder

we use Label Encoder for predicting the output value.
which means we use for 'y'.

```python
label_object = LabelEncoder()
```

```python
y_train_transform=label_object.fit_transform(y_train)
y_test_transform=label_object.transform(y_test)
```

## 1. Applying Ordinal Encoder

we use Ordinal Encoder for Training rhe  values or input variables.
which means we use for 'X'.

```python
ordinalEncoder_object = OrdinalEncoder()
```

```python
X_train_transform=ordinalEncoder_object.fit_transform(X_train)
X_test_transform=ordinalEncoder_object.transform(X_test)
```
