```python
import seaborn as sns
import pandas as pd
import cufflinks as cf
from plotly.offline import iplot
```

```python
# tips=sns.load_dataset('tips')
# tips.head()
tips = sns.load_dataset('tips')
tips
```

```python
# tips['total_bill'].iplot()
import plotly.express as px

fig = px.line(tips, y='total_bill')
fig.show()
```

```python
# tips.groupby('day')['tip'].mean().iplot(kind='bar')
import plotly.express as px

mean_tip_per_day = tips.groupby('day')['tip'].mean().reset_index()
fig = px.bar(mean_tip_per_day, x='day', y='tip')
fig.show()
```

```python
# tips.iplot(kind='scatter',x ='total_bill',y = 'tip',mode = 'markers')
import plotly.express as px

fig = px.scatter(tips, x='total_bill', y='tip')
fig.show()
```

```python
# tips[['total_bill','day']].iplot(kind = 'box',x = 'total_bill',y = 'day')
import plotly.express as px

fig = px.box(tips, x='day', y='total_bill')
fig.show()
```
