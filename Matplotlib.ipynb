```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1) Matplotlib Basics

```python
x = np.linspace(0,5,11) # Linear Spacing(linspace); [I want 11 elements, between 0-5]
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

## 2) Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r') # 'r' --> 'red"
plt.subplot(2,2,2)
plt.plot(y,x,'b') # 'b' --> 'blue'
plt.subplot(2,2,3)
plt.plot(x,x,'g') # 'g' --> 'green'
plt.subplot(2,2,4)
plt.plot(y,y,'y') # 'y' --> 'yellow'
```

## 3) Object Oriented way

## Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

## figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,8), dpi=100) # by default dpi(dots-per-inches) = 100
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.2,0.2])
axis2.plot(x,y)
```

## 4) Types of Plots
###  i) Scatter Plot
### ii) Histogram
### iii) Box Plot

```python
# Scatter Plot
plt.scatter(x,y)
```

```python
# Histogram
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
# Box Plots
plt.boxplot(x)
```

## 5) Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basic.png")
```

## 6) Working with Images

```python
import matplotlib.image as mpimage
img = mpimage.imread("images.jpeg")
# img
plt.imshow(img)
plt.axis("off") # to remove the axis.
```

```python
cropped_img = img[100:200,200:300]
plt.imshow(cropped_img)
plt.axis("off")
```
