# Handle the missing values

```python
import numpy as np
import pandas as pd
```

### Finding the missing Data

#### To Find the missing values we use "isna()".

```python
# np.nan ===>>> Missing Data
# np.nan ===>>> not a number

data = {
    'A':[1,2,np.nan,4,5],
    'B':[1,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
# To find the missing values we use "isna()"
# Here 'True' means missing value.

df.isna()
```

```python
#  To see in which row has how many null values.
df.isna().sum()
```

```python
# To check, is there any null values or empty columns.
df.isna().any()
```

### Removing Missing Data
#### Remember when we remove data it works baasis on the Rows
#### To remove the missing value we use "dropna()".
#### also we can use "drop(thresh=2)" Here thresh means "Atleast how many NOT NULL Values are present."

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3) # --->>> Here "thresh=3" means "There are 3 NOT NULL Values are present."
```

### Filtering Mising Data
#### To filter data wev use "fillna()"

```python
df.fillna(0)
```

```python
val = {'A':0,'B':100,'C':2000,'D':30000}
df.fillna(value=val)
```

```python
df.fillna(df.mean())
```
