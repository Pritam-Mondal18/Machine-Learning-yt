```python
import numpy as np
import pandas as pd
```

# Marging 2 dataframes

```python
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['John','Anna','Peter','Linda','Bob'],
    'depertment':['HR','IT','Finance','IT','HR']
})
# DataFrame 2: Salaries Informations
salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[60000,80000,65000,70000,90000],
    'bonus':[5000,10000,7000,8000,12000]
})
```

```python
print(employees)
print()
print(salaries)
```

```python
pd.merge(employees,salaries) # ----------->> merge two DataFrames (employees,salaries)
```

```python
pd.merge(employees,salaries,on='employee_id') # ------>> on='employee_id' means on the basisof employees_id merge two DataFrames
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner') # ------>> how='inner' means it will concider the inner columns from the DataFrames.
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer') # ------>>> how='outer' means it will consider the outer columns from the DataFrames.
```

```python
pd.merge(employees,salaries,on='employee_id',how='left') # ------>>> here 'left' means basis of employees merge dataframes because employees is in the left side with respect to the salaries.
```

```python
pd.merge(employees,salaries,on='employee_id',how='right') # ------>>> here 'right' means basis of employees merge dataframes because salaries is in the right side with respect to the employees.
```

# Concatination of 2 DataFrame

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3']
})
df2 = pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7']
})
```

```python
print(df1)
print()
print(df2)
```

```python
pd.concat([df1,df2], axis=1)
```

# Jioning of two DataFrame

```python
df1 = pd.DataFrame({
    'name':['Alice','Bob','Charlie']
},index=[1,2,3])
df2 = pd.DataFrame({
    'score':[85,90,75]
},index=[2,3,4])
```

```python
df1.join(df2,how='outer')
```
