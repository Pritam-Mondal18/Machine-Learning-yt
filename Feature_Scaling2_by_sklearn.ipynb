```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df =pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
```

```python
newDf=df.drop('User ID',axis=1)
```

```python
print(newDf)
```

```python
newDf.info()
```

```python
X=newDf.drop(['Purchased'],axis=1)
y=newDf['Purchased']
```

```python
X
```

#Train and Split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

# Applying Standard Scaler

```python
# scaler = MinMaxScaler()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(data=X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(data=X_test_scaled,columns=X_train.columns)
```

```python
X_test_scaled
```

```python
sns.scatterplot(data=X_train_scaled,x='Age',y='EstimatedSalary')
plt.title('After Scaling')
```

```python
sns.scatterplot(data=X_train,x='Age',y='EstimatedSalary',color='red')
plt.title('Original Data')
```
