# `Seaborn`
## We use distribution plots to visualize the distribution of quantitative data.

 ## Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

## :) distplot/histplot
## :) jointplot
## :) pairplot
## :) rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

## **:) distplot/histplot**

```python
sns.displot(df['total_bill'],bins=20,kde=True) # by default bins=20; KDE=Kernel Density Estimeter, it helps to find the density or skewness of the bar chart.
#or
# sns.histplot(df['total_bill'])
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
# sns.displot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
# sns.displot(df['tip'],bins=20,kde=True)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind='scatter')
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind='reg')
```

## :) pairplot

```python
sns.pairplot(df,hue='size',palette='coolwarm')
```

## :) rugplot

```python
sns.rugplot(df)
```
