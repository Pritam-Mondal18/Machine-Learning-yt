# find for a Single 'div' to analysis the data

```python
company = soup.find_all('div', class_='individual_internship')
company_name = []
company_locations = []
company_salary = []
company_experience =[]
job_skills = []
for i in company:
  name_tag = i.find('p', class_='company-name')
  location_tag = i.find ('p', class_ = 'locations')
  salary_tag = i.find('span', class_ = "mobile")
  experience_icon_tag = i.find('i',class_='ic-16-briefcase') # for the extract "Experience" there is no Class so I found the icon class, based on the icon class, found the "experience" section.
  skills_tag = i.find('div',class_='job_skills')
  if name_tag:
    company_name.append(name_tag.text.strip())
  if location_tag:
    company_locations.append(location_tag.text.strip())
  if salary_tag:
    company_salary.append(salary_tag.text.strip())
  if experience_icon_tag:
    experience_tag = experience_icon_tag.find_next('span')
    if experience_tag:
     company_experience.append(experience_tag.text.strip())
  if skills_tag:
    job_skills.append(skills_tag.text.strip())
```

# Creating DataFrame

```python
dic = {'Company Name':company_name,'Locations':company_locations,'Salary':company_salary,'Experience':company_experience,'Skills':job_skills}
df = pd.DataFrame(dic)
df
```

```python
final=pd.DataFrame()

for j in range(1,180):
  url = 'https://internshala.com/jobs/'
  header={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get(url,headers=header, timeout=15).text

  soup = BeautifulSoup(webpage, 'lxml')

  company = soup.find_all('div', class_='individual_internship')
  company_name = []
  company_locations = []
  company_salary = []
  company_experience =[]
  job_skills = []
  for i in company:
    name_tag = i.find('p', class_='company-name')
    location_tag = i.find ('p', class_ = 'locations')
    salary_tag = i.find('span', class_ = "mobile")
    experience_icon_tag = i.find('i',class_='ic-16-briefcase')
    skills_tag = i.find('div',class_='job_skills')
    if name_tag:
      company_name.append(name_tag.text.strip())
    if location_tag:
      company_locations.append(location_tag.text.strip())
    if salary_tag:
      company_salary.append(salary_tag.text.strip())
    if experience_icon_tag:
      experience_tag = experience_icon_tag.find_next('span')
      if experience_tag:
        company_experience.append(experience_tag.text.strip())
    if skills_tag:
      job_skills.append(skills_tag.text.strip())


  dic = {'Company Name':company_name,'Locations':company_locations,'Salary':company_salary,'Experience':company_experience,'Skills':job_skills}
  df = pd.DataFrame(dic)
  # final=final.append(df)
  final = pd.concat([final, df], ignore_index=True)
```
