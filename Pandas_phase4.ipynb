```python
import numpy as np
import pandas as pd
```

# Group By Aggregaton

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
# Group by category and calculate the sum of sales.
df.groupby('Category')['Sales'].sum()
```

```python
# Group by the store and calculate the sum of sales
df.groupby('Store')['Sales'].sum()
```

```python
# Group by multiple columns
# Group by Catagory and Store
pr = df.groupby(['Category','Store'])['Sales'].sum()
pr
```

# Aggregation

```python
df['Sales'].mode()
"""mean,median,mode,min,max,count,std"""
```

```python
df['Sales'].agg(['mean','median','min','max','count','std'])
```
