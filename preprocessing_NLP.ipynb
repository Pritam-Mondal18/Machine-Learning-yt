```python
import pandas as pd
```

```python
df = pd.read_csv("IMDB_cleaned.csv", engine='python', on_bad_lines='skip')
```

```python
new_df = df.drop(columns=["cleaned_review"], axis=1)
```

```python
new_df['review']=new_df['review'].str.lower()
```

```python
new_df.head(5)
```

```python
# remove html tags using regular expressition.
import re
def remove_html_tags(text):
  pattern = re.compile("<.*?>")
  return pattern.sub(r'',text)
```

```python
new_df['review'] = new_df['review'].apply(remove_html_tags)
```

```python
new_df.head()
```

```python
#remove url
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

```python
new_df['review']=new_df['review'].apply(remove_url)
```

```python
import string,time
exclude=string.punctuation
```

```python
exclude
```

```python
#remove punctuation
def remove_punc(text):
  for char in exclude:
    text = text.replace(char,"")
  return text
```

```python
new_df['review']=new_df['review'].apply(remove_punc)
```

```python
new_df['review']
```

```python
start = time.time()
print(remove_punc(new_df['review'][3]))
time1 = time.time() - start
print(time1 * 50000)
```

```python
#remove punctution
def remove_punc1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
start = time.time()
print(remove_punc1(new_df['review'][3]))
time1 = time.time() - start
print(time1 * 50000)
```

```python
from nltk.corpus import stopwords

def remove_stop_words(text):
  new_txt = []
  english_stopwords = stopwords.words('english') # Get the list of English stopwords
  for words in text.split():
    if words not in english_stopwords: # Only append if the word is NOT a stop word
      new_txt.append(words) # Append words correctly
  return " ".join(new_txt) # Return the filtered list
```

```python
# handeling emoji
import re

def remove_emo(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
    "]+", flags=re.UNICODE)

    return emoji_pattern.sub('', text)
```

```python
remove_emo("I am 😀")
```
